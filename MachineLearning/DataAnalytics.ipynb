```python
# Pandas is used to analyze data.
# The name "Pandas" has a reference to both "Panel Data", and "Python Data Analysis" 
# as pd :- refrence or short form for pandas
import pandas as pd

# read_csv("file path") :- read cvs type files
ipl = pd.read_csv("C:\\Users\\SAKSHI BADOLA\\Documents\\MachineLearning\\DataSets\\IPL IMB381IPL2013.csv")

print("ipl type is",type(ipl),"\n\n")

# DataFrame in the pandas library for Python is a two-dimensional(2D), mutable, tabular data structure with labeled axes (rows and columns)
print( pd.DataFrame(ipl))
```

```python
# The head() method returns a specified number of rows, string from the top.
# The head() method returns the first 5 rows if a number is not specified.

print(ipl.head(5))

# to_string () method provide whole file data in to string format
print("\n\n"+ipl.to_string())
```

```python
# .info() is a method of pandas DataFrame that gives a summary of your data.
# It’s very useful to quickly understand your dataset.
# Number of rows and columns
# Column names
# Non-null values in each column (helps detect missing data)
# Data type of each column (int, float, object, etc.)
# Memory usage of the DataFrame

print("info",ipl.info())
```

```python
# .tail() is a pandas method that shows the last few rows of a DataFrame.
# By default, it shows the last 5 rows, but you can change the number.

print("tail", ipl.tail())
```

```python
# .shape tells you the size of a DataFrame (rows × columns).
# It returns a tuple: (number_of_rows, number_of_columns)

print("shape",ipl.shape)
```

```python
# ipl['PLAYER NAME'] Here, ipl is a pandas DataFrame.
#    ipl['PLAYER NAME'] selects the column named "PLAYER NAME".
#    The result is a pandas Series containing all player names.

# [0:5]
#     This is Python slicing.
#     It takes the first 5 rows (index 0 to 4) from that column.
#     print("pd data", ...)
#     It prints the text "pd data" followed by those 5 player names.
# SElecting columns by name
print("pd data",ipl['PLAYER NAME'][0:5])

print("\n\n")

# Selecting columns by name
print("pd data",ipl[['PLAYER NAME','BASE PRICE']][0:5])
```

```python
# Selecting rows and columns by index

# 1. Using .iloc[] → Index-based selection (position numbers)
# Works like normal Python indexing (0-based).
# Syntax: df.iloc[row_index, column_index]

# First row, first column
print(ipl.iloc[0, 0])

# First 5 rows, first 2 columns
print("\n\n",ipl.iloc[0:5, 0:2])

# Specific rows and columns by index list
print("\n\n",ipl.iloc[[0, 2, 4], [0, 1]])
```

<h4>2. Using .loc[] → Label-based selection (row/column names)</h4>
<li>Works with row labels (index values) and column names.</li>
<li><b>Syntax:</b> df.loc[row_label, column_label]</li>

```python
# First 5 rows of PLAYER NAME column
print(ipl.loc[0:4, "PLAYER NAME"])

# Select multiple columns by name
print("\n\n",ipl.loc[0:4, ["PLAYER NAME", "TEAM"]])

# Select specific rows (by index) and specific columns
print("\n\n",ipl.loc[[0, 2, 4], ["PLAYER NAME", "TEAM"]])
```

```python
# 👉 It makes a table that shows:
# It creates a frequency table (cross-tabulation) between two or more categorical variables.
# On the left side (rows) → Ages of players
# On the top side (columns) → Playing roles
#Inside the table → How many players of that AGE have that PLAYING ROLE

pd.crosstab(ipl['AGE'], ipl['PLAYING ROLE'])
```

```python
#sorting
ipl[['PLAYER NAME','SOLD PRICE']].sort_values('SOLD PRICE')[0:5]
```

```python
#sorting
ipl[['PLAYER NAME','SOLD PRICE']].sort_values('SOLD PRICE',ascending = False)[0:5]
```

```python
# 🔹 sorting

# sort_values() → sorts DataFrame by column values
# sort_index() → sorts DataFrame by row/column index
# Can sort ascending/descending
# Can sort by multiple columns with different orders
# Use [0:5] or .head(5) after sorting to get top 5

#🔹 1. Sort by Column Value
ipl.sort_values(by="SOLD PRICE")
```

```python
ipl.sort_values(by="SOLD PRICE", ascending=False)
```

```python
# 🔹 2. Sort by Multiple Columns
ipl.sort_values(by=["TEAM", "SOLD PRICE"], ascending=[True, False])
```

```python
# 🔹 3. Sort by Index
ipl.sort_index()
```

```python
ipl.sort_index(ascending=False)
```

```python
#👉 Sort columns alphabetically:
ipl.sort_index(axis=1)
```

```python
# 🔹 4. With Slicing / Head

ipl[['PLAYER NAME','SOLD PRICE']].sort_values('SOLD PRICE').head(5)
```

```python
# 👉 Shows 5 most expensive players
ipl[['PLAYER NAME','SOLD PRICE']].sort_values('SOLD PRICE', ascending=False).head(5)
```

| Function                       | What it does                     |
| ------------------------------ | -------------------------------- |
| `sort_values("col")`           | Sort by column values            |
| `sort_values(["col1","col2"])` | Sort by multiple columns         |
| `sort_index()`                 | Sort rows by index               |
| `sort_index(axis=1)`           | Sort columns by name             |
| `ascending=True/False`         | Order (low → high or high → low) |
| `.head(5)` or `[0:5]`          | Get top 5 rows after sorting     |
