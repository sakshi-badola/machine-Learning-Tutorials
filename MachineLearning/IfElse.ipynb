```python
# Write a program to calculate electricity bill:
# First 100 units → ₹5 per unit
# Next 100 units → ₹8 per unit
# Above 200 units → ₹10 per unit

unit = int(input("Enter electricity bill"))

if unit <= 100:
  bill = bill * 5
elif unit <= 200:
    bill = 100 * 5 +(unit -100)*8
else :
    bill = 100 * 5 + 100*8 +(unit -200)*10

print("Total Electricity Bill: ₹", bill)
```

```python
# Write a program to classify a triangle as Equilateral, Isosceles, or Scalene using sides.
a, b, c = map(float, input("Enter three side lengths: ").split())

if a + b <= c or a + c <= b or b + c <= a or min(a, b, c) <= 0:
    print("Not a valid triangle")
elif a == b == c :
    print("Equilateral triangle")
elif a == b or b == c or a == c:
    print("Isosceles triangle")
else:
    print("Scalene triangle")
```

```python
# Check whether a number is Armstrong number or not.

num = int(input("Enter number"))
temp = num
sum=0
while num > 0 :
    rem = num %10
    sum = sum + rem*rem*rem
    num //= 10
if temp == sum :
    print("Armstrong number")
else :
    print("Not a Armstrong number")
```
