<h3> List</h3>

```python
#mutables
list = ['Rohit','Dhawan','kholi']

#access element by index 
print(list[0])

#Slice
print("form 0 to 2 :- ",list[0:2]," last index : ",list[-1],end="")


#len
print("\nlist length :- ",len(list),end="")

#concating lists
bol = ['Bumrah','Shami']
final = list + bol

print(final)

# in checks elemet is present or not
'Rohit' in final

#index
final.index('Rohit')
```

<h3>Set</h3



```python
set1={1,3,6,5,8,4,28}
set2= {1,3,10,30}
print("union :- " ,set1.union(set2))
print("intersection :- " ,set1.intersection(set2))
print("difference :- " ,set1.difference(set2))
```

<h3>Dictionary</h3>

```python
wcWinner = {2011:"India",
          1975: "West Indies",
          1983: "India"}

print(set(wcWinner.values()))
print(wcWinner[2011])
print(wcWinner.keys())
print(wcWinner.values())
```
